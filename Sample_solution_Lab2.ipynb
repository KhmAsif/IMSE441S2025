Question 4

```python
car1=car_price.iloc[:200]
```

```python
fig, (ax1, ax2) = plt.subplots(nrows=1,ncols=2, figsize=(12,6))
ax1.plot()
ax2.plot()

ax1.boxplot(car1['Price'])
ax1.set_xlabel('Car Price')
ax1.set_ylabel('Price')


ax2.boxplot(car1['Mileage'])
ax2.set_xlabel('Car Mileage')
ax2.set_ylabel('Mileage')

plt.show()
```

**Interpretation:**
The car price boxplot indicates that most of the used cars are valued between 6500 and 11000. The median used car price is close to $9000. There are no outliers for the 1st 200 datapoints
